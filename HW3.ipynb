### Question 01 

```python
%matplotlib inline
import numpy as np
import math
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import pandas as pd

x1=list(np.random.uniform(-0.5,0.5,50))
y1=list(np.random.uniform(-0.5,0.5,50))
x3=np.ones(100)

t=list(np.zeros(50))

deg=np.random.uniform(0,360,50)
r=np.random.uniform(3,4,50)

for i in range(0,len(deg)):
    degInRad=(deg[i]*math.pi)/180
    
    r_ran=np.random.choice(r)
    x=r_ran*math.cos(degInRad)
    y=r_ran*math.sin(degInRad)
     
    x1.append(x)
    y1.append(y)
    t.append(1)
t=np.array(t).reshape(100,1)
x=np.array([list(a) for a in zip(x1, y1, x3)])

model=tf.keras.Sequential([
    keras.layers.Dense(units=3, input_shape=[3], activation=tf.nn.relu),
    keras.layers.Dense(units=4, activation=tf.nn.relu),
    keras.layers.Dense(units=1, activation=tf.nn.sigmoid)])
# model.compile(optimizer='sgd', loss='mean_squared_error')
model.compile(optimizer='RMSProp', loss='mean_squared_error')
history = model.fit(x, t, epochs=1000)

model.predict(x)

loss = history.history['loss']


epochs = range(len(loss))

plt.plot(epochs, loss, 'r', label='Error')

plt.title('Loss Curve')
plt.legend(loc=0)
plt.figure()
plt.show()


```

### when compared to the optimization technique like Gradient Descent Algorithm, Algorithms like Adam and RMSProp took less time converge to the global minimum value and the error is minimal compared to other algorithms. Unlike in HW2, if we are using tensforflow for implementing the logistic regression overall implementation time is lesser and more optimized. Different activation functions allows non linearity which helps to train the data better.

### Question 02 : Linear Regression

```python
import pandas as pd
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data',
                 na_values='?',
                 names=['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin'],
                 comment='\t',
                 sep=' ',
                 skipinitialspace=True
                )

tempData=data.copy()
```

#### Eliminating Null Values and Converting the data to standard units

```python
import numpy as np
data.dropna(subset=['horsepower'],inplace=True)
# data.reset_index(drop=True, inplace=True)

# data['horsepower']=data['horsepower'].fillna(np.mean(data['horsepower']))

data['mpg']=(data['mpg']-np.mean(data['mpg']))/np.std(data['mpg'])
data['displacement']=(data['displacement']-np.mean(data['displacement']))/np.std(data['displacement'])
data['horsepower']=(data['horsepower']-np.mean(data['horsepower']))/np.std(data['horsepower'])
data['weight']=(data['weight']-np.mean(data['weight']))/np.std(data['weight'])
data['acceleration']=(data['acceleration']-np.mean(data['acceleration']))/np.std(data['acceleration'])

carData=data.drop('mpg',axis=1)
mpgData=data['mpg']
data.head()
```

```python
pd.plotting.scatter_matrix(tempData,figsize=(15,12))
plt.show()
```

### Comparing Each Feature with MPG Data ( Without Co Relation)

```python
def CompareFeaturesWithMPG(df):

    fig = plt.figure(figsize=(15,9))

    ax1 = fig.add_subplot(3,3,1)
    ax2 = fig.add_subplot(3,3,2)
    ax3 = fig.add_subplot(3,3,3)
    ax4 = fig.add_subplot(3,3,4)
    ax5 = fig.add_subplot(3,3,5)
    ax6 = fig.add_subplot(3,3,6)
    ax7=fig.add_subplot(3,3,7)
    df.plot("weight", "mpg", kind="scatter", ax=ax1)
    df.plot("acceleration", "mpg", kind="scatter", ax=ax2)
    df.plot("cylinders", "mpg", kind="scatter", ax=ax3)
    df.plot("displacement", "mpg", kind="scatter", ax=ax4)
    df.plot("model_year", "mpg", kind="scatter", ax=ax5)
    df.plot("origin", "mpg", kind="scatter", ax=ax6)
    df.plot("horsepower", "mpg", kind="scatter", ax=ax7)
    plt.tight_layout()
    plt.show()
    
CompareFeaturesWithMPG(tempData)  
```

### Showing Co-Relation between Each Feature and MPG

```python
def correlation( x, y):
    
    return np.average(x * y)

def slope(x, y):
    r = correlation(x, y)
    return r * np.std((y))/np.std((x))

def intercept(x, y):
    a = slope(x, y)
    return np.average((y)) - a*np.average((x))

def compareFeature(feature,ax):
   
    data_slope=slope(data[feature],mpgData)
    data_intercept=intercept(data[feature],mpgData)

    xIn=np.linspace(-3,3,10)
    yOut=(data_slope*xIn)+data_intercept
    plt.scatter(data[feature],mpgData)
    plt.plot(xIn,yOut,'--',c='g')
    plt.xlabel(feature)
    plt.ylabel('mpg')
    
    
fig = plt.figure(figsize=(15,9))
ax1 = fig.add_subplot(3,3,1)
compareFeature('weight',ax1)
ax2 = fig.add_subplot(3,3,2)
compareFeature('acceleration',ax2)
ax3 = fig.add_subplot(3,3,3)
compareFeature('cylinders',ax3)
ax4 = fig.add_subplot(3,3,4)
compareFeature('displacement',ax4)
ax5 = fig.add_subplot(3,3,5)
compareFeature('model_year',ax5)
ax6 = fig.add_subplot(3,3,6)
compareFeature('origin',ax6)
ax7=fig.add_subplot(3,3,7) 
compareFeature('horsepower',ax7)
plt.tight_layout()
plt.show()
```

#### Linear Regression using Tensorflow & Plotting Original MpG Data vs Predicted MPG Data with Error Curve

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
linearModel=tf.keras.Sequential([
    keras.layers.Dense(units=1)
])
linearModel.compile(optimizer='adam',loss='mean_squared_error')
history=linearModel.fit(carData,mpgData,epochs=1000,verbose=1)
predictedMPG=linearModel.predict(carData)
# print(linearModel.predict(carData))
error=history.history['loss']
epochs = range(len(error))

```

```python
ax=plt.figure(figsize=(12,7))
ax.add_subplot(1,2,1)
plt.plot(epochs, error, 'r', label='Error')
plt.title('Loss Curve')
plt.legend(loc=0)
ax.add_subplot(1,2,2)

plt.scatter(mpgData,predictedMPG)
plt.xlabel('Original MPG Data')
plt.ylabel('Predicted MPG Data')
lims=[-3,3]
plt.xlim(lims)
plt.ylim(lims)
plt.plot(lims,lims,'--',c='g')
plt.show()
```
