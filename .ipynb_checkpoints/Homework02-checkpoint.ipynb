```python
%matplotlib inline
import numpy as np
import random
from matplotlib import pyplot as plt
import math

def sigmoid(uVal):
    return (1/(1+(np.exp(-uVal))))

features=3
lr=0.06
errorList=[]

x1=list(np.random.uniform(-0.5,0.5,50))
y1=list(np.random.uniform(-0.5,0.5,50))
x3=np.ones(100)
t=list(np.zeros(50))

deg=np.random.uniform(0,360,50)
r=np.random.uniform(3,4,50)

for i in range(0,len(deg)):
    degInRad=(deg[i]*math.pi)/180
    
    r_ran=np.random.choice(r)
    x=r_ran*math.cos(degInRad)
    y=r_ran*math.sin(degInRad)
     
    x1.append(x)
    y1.append(y)
    t.append(1)
t=np.array(t).reshape(100,1)

def network(features,in_nodes,hid_nodes,out_nodes):
    w1=np.random.rand(features,in_nodes)
    w2=np.random.rand(in_nodes,hid_nodes)
    w3=np.random.rand(hid_nodes,out_nodes)

    b1=np.random.rand(in_nodes)
    b2=np.random.rand(hid_nodes)
    b3=np.random.rand(out_nodes)
    
    
    while True:
        u1=np.array(list(zip(x1,y1,x3))).dot(w1)+b1
        u2=sigmoid(u1).dot(w2)+b2
        u3=sigmoid(u2).dot(w3)+b3
       
        y=sigmoid(u3)
        
        error=np.sum((y-t)**2)
        errorList.append(error)

        dw1=np.random.rand(features,in_nodes)
        dw2=np.random.rand(in_nodes,hid_nodes)
        dw3=np.random.rand(hid_nodes,out_nodes)

        db1=np.random.rand(in_nodes)
        db2=np.random.rand(hid_nodes)
        db3=np.random.rand(out_nodes)
        
        if(math.isnan(error)):
            print('w1',w1)
            print('w2',w2)
            print('w3',w3)
            print('b1',b1)
            print('b2',b2)
            print('b3',b3)
            break;
            
        for i in range(out_nodes):
            db3[i]=y[i]*(1-y[i])*(y[i]-t[i])
        
        for i in range(hid_nodes):
            for j in range(out_nodes):
                
                dw3[i][j]=u3[i]*db3[j]       
                
                

#         db3=y*(1-y)*(y-t)
#         dw3=(u2.T).dot(db3)

#         db2=(u2)*(1-u2)*np.sum(w3*dw3)
#         dw2=(u1.T).dot(db2)

#         db1=(u1)*(1-u1)*np.sum(w2*dw2)
#         dw1=(np.array(list(zip(x1,y1,x3))).T).dot(db1)
        
        w1=w1-lr*dw1
        b1=b1-lr*db1
        w2=w2-lr*dw2
        b2=b2-lr*db2
        w3=w3-lr*dw3
        b3=b3-lr*db3

# while True:
#     totalError=0   
      
#     for i in range(0,len(t)):
#         u=x1[i]*w1+y1[i]*w2+w3*x3[i]+w4
#         y=sigmoid(u)
        
#         error=(t[i]-y)**2
#         totalError=totalError+error
        
#         dw1=lr*(t[i]-y)*y*(1-y)*x1[i]
#         dw2=lr*(t[i]-y)*y*(1-y)*y1[i]
#         dw3=lr*(t[i]-y)*y*(1-y)*x3[i]
#         dw4=lr*(t[i]-y)*y*(1-y)
        
        
#         w1=w1+dw1
#         w2=w2+dw2
#         w3=w3+dw3
#         w4=w4+dw4
        
#     if(totalError<0.0001 ):
#         print('Plane Equation Found.')
#         break;   
                
# fig = plt.figure()    
# ax = plt.axes(projection='3d')   
# for i in range(len(x1)):
#     if(t[i]==0):

#         ax.scatter(x1[i],y1[i],x3[i],color='red')
#     else:
#         ax.scatter(x1[i],y1[i],x3[i],color='green')


# x=np.linspace(-3,3,num=10)
# y=np.linspace(-3,3,num=10)
# x,y=np.meshgrid(x,y)

# z=(-w1*x-w2*y-w4)/w3

# display.display(plt.gcf()) 
# ax.plot_surface(x,y,z,alpha=0.5,color='red')

# display.clear_output(wait=True)

# time.sleep(0.001)


# plt.show()

network(3,3,4,1)
```
