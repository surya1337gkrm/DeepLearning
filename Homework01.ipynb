```python
%matplotlib inline
import numpy as np
import random
from matplotlib import pyplot as plt
import time
from IPython import display

def sigmoid(uVal):
    return (1/(1+(np.exp(-uVal))))

errorList=[]
x1=[0,0,1,1]
x2=[0,1,0,1]
t=[1,1,1,0]



w1=random.randint(-2,2)*random.random()
w2=random.randint(-2,2)*random.random()
w3=random.randint(-2,2)*random.random()
lr=0.9

col=[]
plt.subplot(121) 
plt.title('Linear Regression')
plt.xlabel('x1')
plt.ylabel('x2')
for i in range(0,len(t)):
    if(t[i]>0):
        color='blue'
    else:
        color='red'
    
    plt.scatter(x1[i],x2[i],c=color)    
count=0
while True:
    count+=1
    
    xi=np.linspace(0,1,5)
    yi=[]
    totalError=0    
    
    for i in range(0,len(t)):
        u=x1[i]*w1+x2[i]*w2+w3
        y=sigmoid(u)
        
        error=(t[i]-y)**2
        
        dw1=lr*(t[i]-y)*y*(1-y)*x1[i]
        dw2=lr*(t[i]-y)*y*(1-y)*x2[i]
        dw3=lr*(t[i]-y)*y*(1-y)
        
        
        w1=w1+dw1
        w2=w2+dw2
        w3=w3+dw3
        
        totalError+=error
        errorList.append(totalError)
        if(w2!=0):
            yi=[]
            for i in range(0,len(xi)):
                yi.append((-w1*xi[i]-w3)/w2) 
        
            if(count==5):
               

                plt.plot(xi,yi,linestyle='--',c='g')
                display.display(plt.gcf())
                display.clear_output(wait=True)
                time.sleep(0.00001) 
                count=0

    if(totalError<0.01 ):
        print('Line Equation Found.')
        break;
plt.subplot(122)        
plt.plot(errorList)       
#plt.show()            
   


```

```python
%matplotlib inline
import numpy as np
import random
from matplotlib import pyplot as plt
import math
def sigmoid(uVal):
    return (1/(1+(np.exp(-uVal))))


x1=list(np.random.uniform(-0.5,0.5,50))
y1=list(np.random.uniform(-0.5,0.5,50))
x3=[]
t=list(np.zeros(50))

#Approach01
# x_d2=list(np.random.uniform(-4,-3,10))
# y_d2=list(np.random.uniform(0,4,10))

# for i in range(len(x_d2)) :
#     x=np.random.choice(x_d2)
#     y=np.random.choice(y_d2)
#     r2=(x)**2+(y)**2
#     if(r2>9 and r2<16):
#         print(np.random.choice([x,-x]),np.random.choice([y,-y]))

# #Approach02
# x_ran=list(np.random.uniform(-4,-3,10))
# r_ran=list(np.random.uniform(3,4,10))  

# y_ran=[]
# for i in range(0,len(x_ran)):
#     y=math.sqrt(abs((np.random.choice(r_ran))**2-(np.random.choice(x_ran))**2))
#     y_ran.append(np.random.choice([y,-y]))
# print(x_ran,y_ran)        
    

#Approach 03 : Using Angles
deg=np.random.uniform(0,180,50)
r=np.random.uniform(3,4,50)

for i in range(0,len(deg)):
    degInRad=(deg[i]*math.pi)/180
    
    r_ran=np.random.choice(r)
    x=r_ran*math.cos(degInRad)
    y=r_ran*math.sin(degInRad)
     
    x1.append(np.random.choice([x,-x]))
    y1.append(np.random.choice([y,-y]))
    t.append(1)
    
for i in range(len(x1)):
    x3.append(x1[i]**2+y1[i]**2)

w1=random.randint(-2,2)*random.random()
w2=random.randint(-2,2)*random.random()
w3=random.randint(-2,2)*random.random()
w4=random.randint(-2,2)*random.random()

lr=0.6

while True:
    totalError=0   
      
    for i in range(0,len(t)):
        u=x1[i]*w1+y1[i]*w2+w3*x3[i]+w4
        y=sigmoid(u)
        
        error=(t[i]-y)**2
        totalError=totalError+error
        
        dw1=lr*(t[i]-y)*y*(1-y)*x1[i]
        dw2=lr*(t[i]-y)*y*(1-y)*y1[i]
        dw3=lr*(t[i]-y)*y*(1-y)*x3[i]
        dw4=lr*(t[i]-y)*y*(1-y)
        
        
        w1=w1+dw1
        w2=w2+dw2
        w3=w3+dw3
        w4=w4+dw4
        
    if(totalError<0.0001 ):
        print('Plane Equation Found.')
        break;   
                
fig = plt.figure()    
ax = plt.axes(projection='3d')   
for i in range(len(x1)):
    if(t[i]==0):

        ax.scatter(x1[i],y1[i],x3[i],color='red')
    else:
        ax.scatter(x1[i],y1[i],x3[i],color='green')


x=np.linspace(-3,3,num=10)
y=np.linspace(-3,3,num=10)
x,y=np.meshgrid(x,y)

z=(-w1*x-w2*y-w4)/w3

display.display(plt.gcf()) 
ax.plot_surface(x,y,z,alpha=0.5,color='red')

display.clear_output(wait=True)

time.sleep(0.001)


# plt.show()
```
